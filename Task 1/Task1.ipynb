```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("worldpopulationdata.csv")
```

```python
df.head()
```

```python
pop_2022 = df["2022"].dropna()
```

```python
plt.figure()
plt.hist(pop_2022, bins=20)
plt.xlabel("Population (2022)")
plt.ylabel("Number of Countries")
plt.title("Distribution of World Population by Country (2022)")
plt.show()
```

```python
top10 = df[["Country Name", "2020"]].dropna() \
        .sort_values(by="2020", ascending=False) \
        .head(10)

plt.figure()
plt.bar(top10["Country Name"], top10["2020"])
plt.xlabel("Country")
plt.ylabel("Population (2020)")
plt.title("Top 10 Countries by Population in 2020")
plt.xticks(rotation=45)
plt.show()
```
